```r
library(Seurat)
library(BPCells)
library(ggplot2)
# needs to be set for large dataset analysis
options(future.globals.maxSize = 1e9)
library(SeuratDisk)
library(dplyr)
options(Seurat.object.assay.version = "v5")
```

```r
data <- read.csv('./reference.csv')[, 2:33]
```

```r
adata <- CreateSeuratObject(t(data))
adata
```

```r
colnames(adata) <-  0:9999
adata[["RNA"]]$data <- adata[["RNA"]]$counts
```

```r
num <- c(5000, 2500, 1250, 625, 312, 156)
get_subsample <- function(obj,number){
    obj <- SketchData(
    object = obj,
    ncells = number,
    method = "LeverageScore",
    sketched.assay = "sketch",
    )
    DefaultAssay(obj) <- "sketch"
    return(colnames(obj[['sketch']]))
}
```

```r
dir.create('./sketch_result')
```

```r
for(i in num){
    print(i)
    t <- list()
    samples <- list()
    for(j in c(1:50)){
    timestart<-Sys.time()
    id = get_subsample(adata, i)
    timeend<-Sys.time()
    runningtime<-timeend-timestart
    t <- cbind(t, runningtime)
    samples <- cbind(samples, id)
    }
    write.csv(samples, paste0('sketch_result/sample_id_', i ,'.csv'))
    write.csv(t, paste0('sketch_result/sample_time_', i ,'.csv'))
}
```
